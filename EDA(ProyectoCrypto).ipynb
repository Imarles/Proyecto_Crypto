Funcion que retorna las monedas con mayor capitalizacion del mercado

```python
def get_top_10_market_cap_coins(coin_list):
    coin_data = cg.get_coins_markets(ids = coin_list, vs_currency = "usd", order = "market_cap_desc")
    top_10_coins = {}
    for coin in coin_data[:10]:
        coin_id = coin["id"]
        coin_name = coin["name"]
        coin_symbol = coin["symbol"]
        coin_market_cap = coin["market_cap"]
        top_10_coins[coin_symbol] = {"Name": coin_name, "Market cap": coin_market_cap, "ID": coin_id}

    return top_10_coins
```

```python
top_10_coins = get_top_10_market_cap_coins(coin_list)
for symbol, data in top_10_coins.items():
    print(f"{data['Name']} ({symbol}): Market Cap - {data['Market cap']}")
```

Datos historicos de cada una de las cryptos.

```python
def get_historical_data(coin_id):
    historical_data = cg.get_coin_market_chart_by_id(id = coin_id ,vs_currency ='usd',days ='max')
    prices = historical_data["prices"]
    market_caps = historical_data["market_caps"]
    total_volumes = historical_data["total_volumes"]

    data = []
    for i in range(len(prices)):
        timestamps = pd.to_datetime(prices[i][0], unit = "ms")
        price = prices[i][1]
        market_cap = market_caps[i][1]
        volumen = total_volumes[i][1]
        data.append({'Timestamp': timestamps, 'Price': price, 'Market Cap': market_cap, 'Volume': volumen})

    df = pd.DataFrame(data)
    return df 
```

```python
top_10_coins = get_top_10_market_cap_coins(coin_list)
cryptos_id = top_10_coins["eth"]["ID"]
cryptos_historical_data = get_historical_data(cryptos_id)

# Convertir los valores de Market Cap y Volume a formato legible si son num√©ricos
cryptos_historical_data["Market Cap"] = cryptos_historical_data["Market Cap"].apply(
    lambda x: '{:.2f}'.format(x) if isinstance(x, (int, float)) else x
)
cryptos_historical_data["Volume"] = cryptos_historical_data["Volume"].apply(
    lambda x: '{:.2f}'.format(x) if isinstance(x, (int, float)) else x
)

# Mostrar el DataFrame con los valores formateados
print(cryptos_historical_data.head())
```

```python
def get_historical_data(coin_id):
    historical_data = cg.get_coin_market_chart_by_id(id = coin_id, vs_currency ='usd', days ='max')
    prices = [entry[1] for entry in historical_data['prices']]
    timestamps = [pd.to_datetime(entry[0], unit ='ms') for entry in historical_data['prices']]
    market_caps = [entry[1] for entry in historical_data['market_caps']]
    volumes = [entry[1] for entry in historical_data['total_volumes']]
    df = pd.DataFrame({'Timestamp': timestamps, 'Price': prices, 'Market Cap': market_caps, 'Volume': volumes})
    return df
```

Matrices de correlacion

```python
# Definir las paletas de colores
viridis_palette = sns.color_palette("viridis")
cubehelix_palette = sns.color_palette("cubehelix")

def plot_correlation_matrix(dataframe, coin, color_map):
    # Filtrar los datos solo para la moneda en espec√≠fico:
    coin_df = dataframe[dataframe["Coin"] == coin]

    # Seleccionar solo las columnas num√©ricas para la matriz de correlaci√≥n
    numeric_columns = coin_df.select_dtypes(include=[float, int]).columns
    correlation_matrix = coin_df[numeric_columns].corr()

    # Crear una matriz de correlaci√≥n con la paleta de colores proporcionada
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap=color_map, center=0, linewidths=.5)
    plt.title(f'Matricez de correlaci√≥n por moneda (Variables Num√©ricas) - {coin}')
    plt.show()

# Llamar a la funci√≥n con tu DataFrame 'final_df' y las monedas deseadas
coins_to_plot = ['btc', 'eth', 'usdt', 'ada', 'dot', 'vet', 'sand', 'ape', 'chz', 'dash']
for i, coin in enumerate(coins_to_plot):
    if i % 2 == 0:  # Usar la paleta viridis para monedas pares
        plot_correlation_matrix(crypto_df, coin, viridis_palette)
    else:  # Usar la paleta cubehelix para monedas impares
        plot_correlation_matrix(crypto_df, coin, cubehelix_palette)
```

Grafico de tendencia

```python
def plot_normalized_histogram(dataframe, timestamp):
    # Copiar el DataFrame original para no modificar los datos originales
    df = dataframe.copy()

    # Crear una instancia del escalador MinMaxScaler
    scaler = MinMaxScaler()

    # Seleccionar solo las columnas de precio de cierre para normalizar
    prices = df.pivot(index='Timestamp', columns='Coin', values='Price')
    normalized_prices = scaler.fit_transform(prices)

    # Crear un DataFrame con los precios normalizados
    normalized_df = pd.DataFrame(normalized_prices, columns=prices.columns, index=prices.index)

    # Establecer el estilo de las visualizaciones de Seaborn
    sns.set(style="whitegrid")

    # Crear un histograma de los precios normalizados en el timestamp dado
    plt.figure(figsize=(10, 6)) 
    plt.hist(normalized_df.loc[timestamp], bins=20, alpha=0.7, edgecolor='black')
    plt.title(f'Tendencia de Precios de Cierre Normalizados - Todas las Criptomonedas {timestamp}')
    plt.xlabel('Precio Normalizado')
    plt.ylabel('Frecuencia')
    plt.tight_layout()

    plt.show()


# Debes cambiar esto al timestamp que desees analizar
timestamp_to_plot = '2023'  
plot_normalized_histogram(crypto_df, timestamp_to_plot)
```

Grafico de distribucion

```python
def plot_kde_distribution(dataframe, coin):
    # Filtrar los datos solo para la moneda en espec√≠fico:
    coin_df = dataframe[dataframe["Coin"] == coin]

    # Establecer el estilo de las visualizaciones de Seaborn
    sns.set(style="whitegrid")

    # Crear un KDE Plot para la distribuci√≥n de precios normalizados
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=coin_df["Price"], fill=True, color="blue")
    plt.title(f"Distribuci√≥n de Precios Normalizados por moneda - {coin}")
    plt.xlabel("Precio de cierre normalizado")
    plt.ylabel("Densidad")
    plt.tight_layout()

    plt.show()

 # Cambiar a la moneda que desees analizar btc, ada, vet, etc...
coin_to_plot = "ada" 
plot_kde_distribution(crypto_df, coin_to_plot)
```
